```python
x = df[["Gender","Customer_Type","Type_of_Travel","Class","Age","Flight_Distance","Inflight_entertainment","Baggage_handling","Cleanliness","Departure_Delay_in_Minutes","Arrival_Delay_in_Minutes"]]
y = df[["satisfaction"]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
y_train.shape , y_test.shape , x_train.shape , x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train.values.ravel())
```

```python
model.score(x_train , y_train)
```

```python
model.score(x_test , y_test)
```

```python
import joblib
joblib.dump(model, "customer_model.pkl")
```

```python
new_pred = [[1,	1,	1,	2,	23,	192,	2,	4,	2,	3,	0.0]]
pred = model.predict(new_pred)
print(pred)
```
